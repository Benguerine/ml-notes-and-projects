```python
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12,12))
ax1.plot(X,y)
ax2.plot(X,[2,6,8,9])
ax3.plot([5,9,7,1], y)
ax4.plot([1,2,3,4],[1,4,9,16])

plt.show()
```

```python
import numpy as np
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
ax1.plot(X,np.sqrt(y)**4)
ax2.plot(X,[2,6,8,9])
ax3.plot([5,9,7,1], y)
ax4.plot([1,2,3,4],[1,4,9,16])

plt.show()
```
