```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig('sample.png')
```

```python
# Subplot option1:
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot for each differnent axis
ax1.plot(x,np.array(x)/2); # Convert x to a NumPy array for element-wise division
ax2.plot(x,np.array(x)**2);
ax3.plot(x,np.sqrt(x));
ax4.plot(np.random.randn(1000));
```

```python
# Let's plot from another dataset
path = "/content/drive/MyDrive/Data Scienc/heart-disease.csv"
df = pd.read_csv(path)
df.head(10)
```

```python
df.shape
```

```python
# Creat scatter
df.plot(x = "age", y="trestbps",kind="scatter")
```

```python
df.plot.hist(subplots = True, figsize=(10,30));
```

```python
over_50 = df[df["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind = "scatter",x="age",y="chol",c ="target")
```

```python
over_50.plot(kind = "scatter",x="age",y="chol",c ="target",cmap="plasma")
```

```python
# Using OO methods
# import matplotlib.pyplot as plt
# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the data on the axes
over_50.plot(kind="scatter", x="age", y="chol", c="target", cmap="plasma", ax=ax)

# Display the plot
plt.show()
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"],
                     over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legned
ax.legend(*scatter.legend_elements(),
          title="Target")

# Add a meanline
ax.axhline(over_50["chol"].mean(),
           linestyle="--"); # style the line to make it look nice
```

```python
plt.style.use('default')
# Let's plot some graphs
fig ,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
# Costumize ax0
ax0.set(title="Heart dessise and cholesterol level", xlabel="Age" , ylabel="cholestrol");
# Now add legend to ax0
ax0.legend(*scatter.legend_elements(),title="target");
# Add mean line
ax0.axhline(over_50["chol"].mean() , linestyle="--");


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
# Costumiz ax1
ax1.set(title="Heart dessise and thalach level",xlabel="Age",ylabel="thalach");
# Now add lengent to ax1
ax1.legend(*scatter.legend_elements(),title="target");
# add mean line
ax1.axhline(over_50["thalach"].mean(),linestyle="--");

# Add title to figuer
fig.suptitle("Heart dessise analysis", fontsize=16 , fontweight="bold");
```
