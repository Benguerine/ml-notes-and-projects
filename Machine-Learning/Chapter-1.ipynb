```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

> ðŸ”‘: https://scikit-learn.org/stable/modules/linear_model.html

```python
X_reg, y_reg = make_regression(n_samples=200, n_features=1, noise=15, random_state=42)
```

```python
X_reg
```

```python
y_reg
```

```python
len(X_reg)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
X_train
```

```python
# Models
models_reg = {
    "LinearRegression": LinearRegression(),
    "Ridge": Ridge(alpha=0.1),
    "Lasso": Lasso(alpha=0.1),
    "ElaticNet": ElasticNet(alpha=0.1, l1_ratio=0.9)
}
for cle, value in models_reg.items():
  value.fit(X_train, y_train)
  y_pred = value.predict(X_test)
  mse = mean_squared_error(y_test, y_pred)
  print(f"{cle} Regression - MSE: {mse:.3f}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_test, y_test, color ='gray', label = 'data')
x_line = np.linspace(X_test.min(), X_test.max(), 100).reshape(-1,1)

for cle, value in models_reg.items():
  y_line = value.predict(x_line)
  plt.plot(x_line, y_line, label= value)

plt.title("Linear Regression models with Regularization")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
