```python
import numpy as np
import matplotlib.pyplot as plt #2D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
len(data.data)
```

```python
data
```

```python
X = data.data
y = data.target
```

```python
X, y
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
print(pca.explained_variance_)
```

```python
# 2.93808505 * 100 <--> pc1
# 0.9201649 * 100 <--> pc2
```

```python
y
```

```python
z = np.unique(y)
z
```

```python
plt.figure(figsize=(8,6))
for label in np.unique(y):
  plt.scatter(X_pca[y==label, 0], X_pca[y==label,1],label=data.target_names[label])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.title("PCA")
plt.show()
```

##**t-sne**

```python
# 2-D
from sklearn.datasets import load_iris # dataset
from sklearn.manifold import TSNE # algorithm
import matplotlib.pyplot as plt # plot
import numpy as np # calculate
```

```python
iris = load_iris()
```

```python
iris
```

```python
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
print(iris.feature_names) # 4D --> 2D
```

```python
tsne = TSNE(n_components=3)
X_tsne = tsne.fit_transform(X)
```

```python
X_tsne
```

```python
np.unique(y)
```

```python
plt.figure(figsize=(10,6))
#[0,1,2]
for label in np.unique(y):
  plt.scatter(
      X_tsne[y==label, 0],
      X_tsne[y==label,1],
      label=iris.target_names[label]
  )

plt.xlabel("tsne1")
plt.ylabel("tsne2")
plt.title("t_sne on iris dataset")
plt.legend()
plt.show()

```
