```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

> ðŸ”‘: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html

```python
# Generate data
X,_ = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
# Elbow method
wcss=[]
for k in range(1,11):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_) ## wcss[k] = kmeans.intertia_
```

>ðŸ”‘:https://www.w3schools.com/python/ref_list_append.asp

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of K')
plt.ylabel('WCSS')
plt.show()
```

**`k=4`**

```python
kmeans = KMeans(n_clusters=4)
labels = kmeans.fit_predict(X)
```

```python
X
```

```python
labels # y
```

```python
plt.scatter(X[:,0],X[:,1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='red', marker='x', s=100)
plt.title('K-means Clustering')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
X,_ = make_moons(n_samples=300, noise=0.05, random_state=42)
```

```python
X
```

```python
X = StandardScaler().fit_transform(X)
```

```python
X
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)
```

```python
labels
```

```python
plt.scatter(X[:,0],X[:,1], c=labels, cmap='plasma')
plt.title('DBSCAN Clustering')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Sample data points (2D)
X = np.array([
    [1, 2], #0
    [2, 3], #1
    [5, 6], #2
    [8, 8], #3
    [1, 0], #4
    [7, 5], #5
    [6, 7], #6
    [3, 2] #7
])

# Perform Agglomerative Clustering
cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
labels = cluster.fit_predict(X)

# Plotting the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=100)
plt.title('Agglomerative Clustering with sklearn')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
